```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
#importing csv file

df = pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
#top 10 entries in csv file

df.head(10)
```

```python
#information of all the column with their data type

df.info()
```

```python
#to drop null columns

df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
#to check if null columns are deleted

df.info()
```

```python
#to check null values in the csv file

pd.isnull(df)
```

```python
#total null value in each column

pd.isnull(df).sum()
```

```python
#total number of rows and column before droping the null rows

df.shape
```

```python
#to drop the null rows

df.dropna(inplace=True)
```

```python
#checking if null rows are dropped or not

df.shape
```

```python
#to change the data type of Amount  column from float to interger

df['Amount']= df['Amount'].astype('int')
```

```python
#to check the datatype of Amount Column after altering it from float to Interger

df['Amount'].dtypes
```

<span style="font-size:25px;">**EXPLORATORY DATA ANALYSIS**</span>

<span style="font-size:20px;">**Gender**</span>

```python
#bar graph for Gender count

ax = sns.countplot(x = 'Gender',data = df)
sns.set(rc={'figure.figsize':(2,5)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Gender Count',fontsize=20)
plt.show
```

```python
#bar chart for total sales with respect to gender

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen,hue='Gender',palette=['#FF6347', '#4682B4'])
plt.title('Purchase with respect to Genders',fontsize=20)
plt.show
```

**From above graph we can see that most of the buyers are Females and their purchasing power is greater than males**

<span style="font-size:20px;">**AGE**</span>

```python
#bar graph for genders in age group

ax=sns.countplot(data=df, x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Group by gender',fontsize=20)
plt.show
```

```python
# bar graph of amount spent by age groups

sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group', y='Amount', hue='Age Group',palette=['#FF6347', '#4682B4', '#8A2BE2', '#FF7F50', '#32CD32', '#FFD700', '#7FFF00'] ,data=sales_age, legend=False)
plt.title("Sales by age")
plt.show
```

**From above graph we can see that most of the buyers are  of age group between 26-35 yrs females**

<span style="font-size:20px;">**STATE**</span>

```python
#bar graph for order from top 10 states

state_orders=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=state_orders,x='State',y='Orders',hue='State',palette= ['#FF5733', '#33FF57', '#3357FF', '#F1C40F', '#8E44AD', '#E74C3C', '#1F77B4', '#2CA02C', '#D62728', '#9467BD'])
plt.title("Orders per State",fontsize=20)
plt.show
```

**Uttar Pradesh  have highest volume of order placed and 1oth position we have Gujarat**

```python
#bar graph for sales from top 10 states

state_sales=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=state_sales,x='State',y='Amount',hue='State',palette= ['#FF5733', '#33FF57', '#3357FF', '#F1C40F', '#8E44AD', '#E74C3C', '#1F77B4', '#2CA02C', '#D62728', '#9467BD'])
plt.title("Sales per State",fontsize=20)
plt.show
```

**Uttar Pradesh  have highest sales of order places and 1oth position we have Gujarat**


**From above graphs we can see that most of the orders & total sales are from Uttar Pradesh, Maharashtra and Karnataka respectively**

<span style="font-size:20px;">**Marital Status**</span>

```python
#bargraph for count of married and unmarried customers

ax= sns.countplot(data=df,x='Marital_Status',hue='Marital_Status',palette=['#FF5733', '#33FF57'])
sns.set(rc={'figure.figsize':(9,4)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Marital Status",fontsize=20)
plt.show
```

```python
#bar graph for amount spend by gender based of their marital status

state_sales=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=state_sales,x='Marital_Status',y='Amount',hue='Gender')
plt.title("Amount spend by Gender based on their Marital Status",fontsize=20)
plt.show
```

 **From above graph we can see tha that most of the buyers are married(women) and they have high purchasing power**

<span style="font-size:20px;">**Occupation**</span>

```python
#bar graph for count of each occupation of the customers

sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x="Occupation")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Count of Occupation",fontsize=20)
plt.show
```

```python
#bar graph for sales amount per occupation

occupation_sales=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=occupation_sales,x='Occupation',y='Amount',hue="Occupation",palette=['#FF5733', '#33FF57', '#3357FF', '#F1C40F', '#8E44AD', '#E74C3C', '#1F77B4', '#2CA02C', '#D62728', '#9467BD','#FF69B4', '#4B0082', '#FFD700', '#ADFF2F', '#FF4500'])
plt.title("Sales per  occupation",fontsize=20)
plt.show
            
```

**From above graph we can see that most of the customers are working in IT,Healthcare and Aviation sector**

<span style="font-size:20px;">**Product Category**</span>

```python
#bar graph for product category count

sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x='Product_Category',hue='Product_Category',palette=['#FF5733', '#33FF57', '#3357FF', '#F1C40F', '#8E44AD', '#E74C3C', '#1F77B4', '#2CA02C', '#D62728', '#9467BD','#FF69B4', '#4B0082', '#FFD700', '#ADFF2F', '#FF4500', '#00CED1', '#FF1493', '#4682B4'])

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Product Category Count",fontsize=20)
plt.show
```

```python
#bar graph for sales per product category

sales_product_category=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(22,5)})
sns.barplot(data=sales_product_category,x='Product_Category',y='Amount',hue='Amount',palette=['#FF5733', '#33FF57', '#3357FF', '#F1C40F', '#8E44AD', '#E74C3C', '#1F77B4', '#2CA02C', '#D62728', '#9467BD'])
plt.title('Sales per Product Category',fontsize=20)
plt.show
```

**From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category**

```python
#bar graph for number of orders by product category

product_id_orders=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=product_id_orders,x='Product_ID',y='Orders',hue='Product_ID',palette=['#FF5733', '#33FF57', '#3357FF', '#F1C40F', '#8E44AD', '#E74C3C', '#1F77B4', '#2CA02C', '#D62728', '#9467BD'])
plt.title('Number of orders by Product ID',fontsize=20)
plt.show
```
